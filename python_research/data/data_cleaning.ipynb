```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_bidhan = pd.read_csv("../data/durgapur_bidhannagar_final.csv")
df_bidhan["datetime"] = pd.to_datetime(df_bidhan["datetime"])
df_bidhan
```

```python
df_chandidas = pd.read_csv("../data/durgapur_chandidas_market_final.csv")
df_chandidas["datetime"] = pd.to_datetime(df_chandidas["datetime"])
df_chandidas
```

```python
df_city = pd.read_csv("../data/durgapur_city_centre_final.csv")
df_city["datetime"] = pd.to_datetime(df_city["datetime"])
df_city
```

```python
df_dsp = pd.read_csv("../data/durgapur_dsp_area_final.csv")
df_dsp["datetime"] = pd.to_datetime(df_dsp["datetime"])
df_dsp
```

```python
for df in [df_bidhan, df_chandidas, df_city, df_dsp]:
    print(df.info())
    print(df.describe())
    print(df.isnull().sum())
```

```python
df_cols = {
    0: df_bidhan,
    1: df_chandidas,
    2: df_city,
    3: df_dsp
}
df_list = []
for station_id, df in df_cols.items():
    df = df.copy()
    df.insert(df.columns.get_loc("datetime") + 1, "station_id", station_id)
    df.drop(columns=["lat","long"], inplace=True)
    df_list.append(df)

df_list[0]
```

```python
df_list_sorted = [df.sort_values("datetime") for df in df_list]
final_df = pd.concat(df_list_sorted, ignore_index=True)
final_df
```

```python
final_df.describe()
```

```python
import seaborn as sns

plt.figure(figsize=(12, 8))
sns.heatmap(final_df.drop(columns="station_id").corr(), annot=False, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
idx = final_df.columns.get_loc("datetime")


h = final_df['datetime'].dt.hour
d = final_df['datetime'].dt.dayofyear
m = final_df['datetime'].dt.month


final_df.insert(idx + 1, "hour_sin",  np.sin(2 * np.pi * h / 24))
final_df.insert(idx + 2, "hour_cos",  np.cos(2 * np.pi * h / 24))
final_df.insert(idx + 3, "date_sin",  np.sin(2 * np.pi * d / 31))
final_df.insert(idx + 4, "date_cos",  np.cos(2 * np.pi * d / 31))
final_df.insert(idx + 5, "month_sin", np.sin(2 * np.pi * m / 12))
final_df.insert(idx + 6, "month_cos", np.cos(2 * np.pi * m / 12))
final_df.insert(idx + 7, "year",      final_df['datetime'].dt.year)

sid = final_df.pop("station_id")
final_df.insert(idx + 8, "station_id", sid)

final_df.drop(columns=["datetime"], inplace=True)

final_df
```

```python
final_df.to_csv("durgapur_final.csv",index=False)
```
